```python
fiis = fiis.drop(labels=[0,1])
```

```python
fiis = fiis.fillna(0)
```

```python
def filtro(pv,pa,dy,qa):
    filtrado = fiis.loc[(fiis["P/VP"] <= pv) & (fiis["PREÇO ATUAL (R$)"] <= pa) & (fiis["DY (12M) MÉDIA"] >= dy) & (fiis["QUANT. ATIVOS"] >= qa)]
    return filtrado
```

```python
menor = fiis.loc[(fiis["P/VP"] <= 1.1) & (fiis["PREÇO ATUAL (R$)"] <= 100) & (fiis["DY (12M) MÉDIA"] >= 0.02) & (fiis["QUANT. ATIVOS"] >= 2)]

```

```python
display(menor)
```

```python
menor.to_excel("tabela.xlsx")
```
